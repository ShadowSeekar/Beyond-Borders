```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('data_uni.csv')
data
```

```python
data.describe()
```

```python
data.columns
```

```python
x = data[['CGPA']]
y = [1 if i < 8.085000 else 0 for i in data['CGPA']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
model = DecisionTreeClassifier(random_state=42)
```

```python
model.fit(x_train, y_train)
```

```python
score = model.score(x_test, y_test)
print("Model accuracy: ", score)
```

```python
input_value = 10
prediction = model.predict([[input_value]])
filtered_data = data[data['CGPA'] < input_value]
print(f"Values less than {input_value}: \n{filtered_data}")
```

```python

dfs = pd.DataFrame(filtered_data, columns=['world_rank', 'institution', 'country', 'quality_of_education',
       'quality_of_faculty', 'influence', 'patents', 'Average', 'rank_by_avg',
       'GRE Score', 'TOEFL Score', 'CGPA'])
dfs
```
